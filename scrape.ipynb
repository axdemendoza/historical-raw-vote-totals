```python
url = 'https://en.wikipedia.org/wiki/2016_United_States_presidential_election'
```

```python
df = pd.read_html(url, attrs={"class": "wikitable sortable"})[2]
```

```python
df
```

```python
swing_states = ['Fla.', 'Ga.', 'N.C.', 'Pa.', 'Ohio', 'Wis.', 'Mich.', 'Texas', 'Ariz.']
swing_states_2 = ['Florida', 'Georgia', 'North Carolina', 'Pennsylvania', 'Ohio', 'Wisconsin', 'Michigan', 'Texas', 'Arizona']
```

```python
df.drop(df.tail(1).index, inplace=True)
```

```python
df = df.iloc[:, 0:7]
```

```python
df.drop('Unnamed: 1_level_0', axis=1, level=0, inplace=True)
```

```python
melt_lvl_1 = df.melt(id_vars=['State ordistrict'], col_level=1)
```

```python
melt_lvl_1.columns = ['State', 'variable', 'value']
```

```python
df
```

```python
melt_lvl_0 = df.melt(id_vars=['State ordistrict'], col_level=0)
```

```python
melt_lvl_0
```

```python
melt_lvl_0 .columns = ['State', 'variable', 'value']
```

```python
df = pd.merge(left=melt_lvl_0, right=melt_lvl_1, how='inner', on=['State', 'value'])
```

```python
df['State'] = df['State'].replace('Arizona★', 'Arizona')
```

```python
df.columns = ['State', 'Candidate', 'value', 'variable']
```

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce')
```

```python
df = df.pivot_table(index=['State', 'Candidate'], columns='variable', values='value')
```

```python
df
```

```python
df.drop('EV', axis=1, inplace=True)
```

```python
df = df.reset_index()
```

```python
df['Year'] = '2016'
```

```python
df['State'] = df['State'].replace('Wisconsin[117]', 'Wisconsin')
df['State'] = df['State'].replace('Ohio[116]', 'Ohio')
```

```python
df = df[df['State'].isin(swing_states)]
```

```python
df
```

```python
df.to_csv('2016.csv')
```
